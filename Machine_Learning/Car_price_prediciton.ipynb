```python
aspiration_mapper = {
    'std': 0,
    'turbo': 1
}

data.aspiration = data.aspiration.map(aspiration_mapper)

data
```

```python
data.doornumber.unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
data['doornumber'] = enc.fit_transform(data.doornumber)
data.head()
```

```python
data.carbody.unique()
```

```python
data.carbody = data.carbody.astype('category')
data.carbody = data.carbody.cat.codes
```

```python
data.head()
```

```python
data.drivewheel.unique()
```

```python
from sklearn.preprocessing import LabelBinarizer
enc = LabelBinarizer()
data['drivewheel'] = enc.fit_transform(data['drivewheel'])

data.head()
```

```python
data.enginelocation.unique(

)
```

```python
data_enginelocation_map = {
    'front': 0,
    'rear': 1
}

data['enginelocation'] = data.enginelocation.map(data_enginelocation_map)
```

```python
data.fuelsystem.unique()
```

```python
from sklearn.preprocessing import LabelBinarizer
enc = LabelBinarizer()
data.fuelsystem = enc.fit_transform(data.fuelsystem)
```

```python
data.info()
```

```python
data.cylindernumber = enc.fit_transform(data.cylindernumber)
data.head()
```

```python
data.info()
```

```python
data.enginetype.unique()
```

```python
data.enginetype = data.enginetype.astype('category')
data.enginetype = data.enginetype.cat.codes
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split


features = data.drop('price', axis=1)
target = data['price']

x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.4, random_state = 380)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
prediction = model.predict(x_train)
```

```python
score = r2_score(prediction, y_train)
print(round(score*100, 2))
```
